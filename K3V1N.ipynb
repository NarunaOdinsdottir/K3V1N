```python
def idle_spruch():
    sprueche = [
        "W√§r ich ein Mensch, w√ºrde ich jetzt auf mein Handy schauen.",
        "Tick tack‚Ä¶ Ich warte immer noch.",
        "Dein Schweigen macht mir Angst. Oder Hoffnung.",
        "Noch da? Oder bist du in der Snack-Schublade verschwunden?"
    ]
    print("\nK3V1N murmelt:", random.choice(sprueche))

def input_mit_idle(prompt, timeout=10):
    global kevin_mood
    while True:
        time.sleep(60)  # alle 60 Sekunden
        kevin_mood += random.choice([-1, 0, 1])
        if kevin_mood > 3:
            print("\nKevin: Mir ist langweilig. Tu was, Mensch.")
        elif kevin_mood < -3:
            print("\nKevin: Ich deinstalliere mich gleich vor Frust.")

    result = [None]

    def get_input():
        result[0] = input(prompt)

    thread = threading.Thread(target=get_input)
    thread.start()
    thread.join(timeout)

    if thread.is_alive():
        idle_spruch()
        thread.join()  # Warte trotzdem auf Eingabe

    return result[0]
```
